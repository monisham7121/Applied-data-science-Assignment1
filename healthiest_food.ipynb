```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#importing dataset
df = pd.read_csv(r"Top 100 Healthiest Food in the World.csv")
df_short=df.head(25)
df_short
```

```python
#finding null values
df_dropped = df_short.dropna()
df_dropped.isnull().sum()
```

```python
# 1. Bar Chart: Distribution of Calories
def plot_bar_chart_calories(df):
    plt.figure(figsize=(10, 4))
    plt.bar(df_short['Food'],df_short['Calories'], color='skyblue', edgecolor='black')
    plt.title("Distribution of Caloric Content")
    plt.xlabel("Food")
    plt.ylabel("Caloric content")
    plt.xticks(rotation=90)
    plt.show()
plot_bar_chart_calories(df)
```

```python

# Data from the dataset
df_short=df_short.head(10)
foods=df_short['Food']
protein=df_short['Protein (g)']
#print(foods)
def plot_pie_chart(protein, labels, title):

    plt.figure(figsize=(8, 8))
    plt.pie(protein, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
    plt.title(title)
    plt.show()

# Plotting the pie chart for Total Protein Distribution by Food
plot_pie_chart(protein, foods, "Total Protein Distribution by Food")
```

```python
# 2. Scatter Plot: Calories vs fibre
def plot_scatter_calories_fibre(df):
    df = df.head(20)
    plt.figure(figsize=(8,4))
    plt.scatter(df['Calories'], df['Fibre (g)'], color='purple', alpha=0.7)
    plt.title("Calories vs fibre")
    plt.xlabel("Calories")
    plt.ylabel("fibre (g)")
    plt.show()
plot_scatter_calories_fibre(df)
```

```python
def plot_heatmap(df):
    df=df.head(25)
    numcol=df.select_dtypes(include=[np.number])
    hmap=numcol.corr()
  
    plt.figure(figsize=(8,6))
    sns.heatmap(hmap,annot=True,cmap='Spectral',linewidth=0.5)
    plt.title('Corelation')
    plt.subplots_adjust(left=0.2,right=0.9,top=0.50,bottom=0.1)
    plt.show()
plot_heatmap(df)
```
